```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

plt.style.use('seaborn')

def fourier_onda_quadrada(x,n):
  m = 2*n-1
  #m = 2*n
  f = (4/np.pi)*(1/m)*np.sin(m*np.pi*x/L)
  #f = (4/np.pi)*(1/m)*np.cos(m*np.pi*x/L)
  return f

L = np.pi
ciclos = 1
x = np.linspace(-0.001, 2*ciclos*L, 1000)

f = 0
n = 1
n_total = 5

while n<5:
  f += fourier_onda_quadrada(x,n)
  plt.plot(x,f,label="n = {}" .format(2*n-1))
  #plt.plot(x,f,label="n = {}" .format(2*n))
  n +=1

plt.plot(x,signal.square(x), color= 'k')
plt.legend()
plt.show()
```

**1)** Alterando o valor de m para receber apenas os nmeros pares e mantendo a funo f definida previamente como a onda quadrada representada pelo seno.

> m = 2 * n-1 (apenas mpares)

Ficou,

> m = 2 * n (apenas pares)

E alm dessa alterao fiz a seguinte mudana:

Antes:
> plt.plot(x,f,label="n {}" .format(2*n-1))

E depois:
> plt.plot(x,f,label="n {}" .format(2*n))

Com as alteraes obtive algo parecido com o grfico mostrado em sala porm o sinal foi "espelhado", diferente do apresentado em sala que tratava de um sinal cosseno (par).
